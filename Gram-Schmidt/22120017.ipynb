## Bài 1) Cho A là ma trận có thể phân rã QR. Sinh viên viết chương trình in ra ma trận Q và R, biết rằng A = QR.


```python
import numpy as np
import sys
```

```python
# Hàm tính norm của vector
def norm(v):
    return np.sqrt(np.dot(v, v))

# Hàm chuyển đổi vectơ thành vectơ đơn vị
def normalize(v):
    return v / norm(v)

# Hàm tính inner product của hai vector
def inner_product(u, v):
    return np.dot(u, v)

# Nhập ma trận A
n = int(input('Nhập số hàng của ma trận A: '))
m = int(input('Nhập số cột của ma trận A: '))
A = np.zeros((n, m))
print('Nhập ma trận A[' + str(n) + 'x' + str(m) + ']:')
for i in range(n):
    for j in range(m):
        A[i][j] = float(input('a[' + str(i) + '][' + str(j) + ']='))

print(A)
np.set_printoptions(precision=4, suppress=True)

# Khởi tạo ma trận Q và R
Q = np.zeros((n, m))
R = np.zeros((m, m))

# Phân tích ma trận A thành dạng QR bằng phương pháp Gram-Schmidt
for j in range(m):
    v = A[:, j]
    for i in range(j):
        R[i, j] = inner_product(Q[:, i], A[:, j])
        v = v - R[i, j] * Q[:, i]
    R[j, j] = norm(v)
    Q[:, j] = normalize(v)

print('Ma trận Q:')
print(Q)
print('Ma trận R:')
print(R)
```


**Ý tưởng thực hiện phân tích Gram-Schmidt:**
1. **Nhập ma trận A:** Người dùng nhập số hàng và số cột của ma trận A, sau đó nhập giá trị cho từng phần tử của ma trận A.
2. **Khởi tạo ma trận Q và R:** Tạo hai ma trận Q và R ban đầu, mỗi ma trận có kích thước phù hợp với ma trận A và tất cả các phần tử đều là 0.
3. **Phân tích Gram-Schmidt:**
   - Lặp qua từng cột của ma trận A.
     - Bắt đầu với vectơ cột đầu tiên của A, gán nó vào vectơ cột đầu tiên của Q.
     - Đối với mỗi vectơ cột tiếp theo:
       - Tính phần chiếu của vectơ cột đang xét lên các vectơ cột đã chọn trước đó trong Q.
       - Trừ đi phần chiếu khỏi vectơ cột gốc để tạo ra một vectơ mới.
       - Chuẩn hóa vectơ mới thành một vectơ đơn vị.
       - Gán vectơ đơn vị mới vào ma trận Q.
       - Tính các hệ số của vectơ đơn vị mới để cập nhật ma trận R.
4. **Kết quả:** Ma trận Q sẽ chứa các vectơ đơn vị trực giao và ma trận R sẽ chứa các hệ số của phép biến đổi để chuyển đổi A thành QR.


## Bài 2) Mở rộng:
- Tìm hiểu hàm/ phương thức tương ứng của các thư viện và thực hiện nó, so sánh kết 
quả.
- Tìm hiểu và trình bày ứng dụng của QR decomposition

```python
# Nhập ma trận A
n = int(input('Nhập số hàng của ma trận A: '))
m = int(input('Nhập số cột của ma trận A: '))
A = np.zeros((n, m))
print('Nhập ma trận A[' + str(n) + 'x' + str(m) + ']:')
for i in range(n):
    for j in range(m):
        A[i][j] = float(input('a[' + str(i) + '][' + str(j) + ']='))

print(A)
np.set_printoptions(precision=4, suppress=True)

# Phân tích QR của ma trận A
Q, R = np.linalg.qr(A)

print('Ma trận trực giao Q là:')
print(Q)
print('Ma trận tam giác trên R là:')
print(R)
```

1. Tìm hiểu hàm/ phương thức tương ứng của các thư viện và thực hiện nó, so sánh kết quả.
- NumPy cung cấp hàm numpy.linalg.qr để thực hiện phân tích QR của một ma trận. 
- Kết quả của việc giải thuần túy và dùng thư viện có sẵn cho ra kết quả khác nhau cụ thể là dấu của ma trận
2. Tìm hiểu và trình bày ứng dụng của QR decomposition:
Phân tích QR (QR decomposition) là một phương pháp toán học để phân tích một ma trận thành tích của hai ma trận: một ma trận Orthogonal (Q) và một ma trận tam giác trên (R). Phân tích này rất hữu ích trong nhiều ứng dụng khác nhau, bao gồm:

- Giải hệ phương trình tuyến tính: QR decomposition có thể được sử dụng để giải hệ phương trình tuyến tính. Bằng cách phân tích ma trận hệ số của hệ phương trình thành tích của ma trận Orthogonal và ma trận tam giác trên, ta có thể dễ dàng giải quyết hệ phương trình này.
- Phân tích tuyến tính: Trong định lý khoảng cách gần nhất tuyến tính (NNLS), QR decomposition được sử dụng để giải quyết vấn đề tối ưu hóa có điều kiện ràng buộc là các biến phải là số không âm.
- Tính toán giá trị suy biến: QR decomposition có thể được sử dụng để tính toán giá trị suy biến của một ma trận.
- Tính toán giá trị riêng và vector riêng: Trong một số trường hợp, QR decomposition có thể được sử dụng để tính toán giá trị riêng và vector riêng của một ma trận.
- Ứng dụng trong các thuật toán tối ưu hóa: QR decomposition cũng được sử dụng trong các thuật toán tối ưu hóa để giảm độ phức tạp tính toán.
- Ứng dụng trong phân tích dữ liệu: Trong xử lý tín hiệu và phân tích dữ liệu, QR decomposition được sử dụng để giảm độ phức tạp tính toán và cải thiện hiệu suất.
